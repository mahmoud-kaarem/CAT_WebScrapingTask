```python
import requests
from bs4 import BeautifulSoup
from itertools import zip_longest
import csv
import re
```

```python
developer = []
type = []
platform = []
stock_Status = []
game_name = []
game_description = []
genre = []
price = []
links = []
```

```python
page_num = 1
```

```python
while page_num<=94:
    web_result = requests.get(f"https://sandbox.oxylabs.io/products?page={page_num}")
    pre_url = "https://sandbox.oxylabs.io"
    src = web_result.content
    soup = BeautifulSoup(src,"lxml")
    games_names = soup.find_all("h4",{"class":"title css-7u5e79 eag3qlw7"})
    genres = soup.find_all("p",{"class":"category css-8fdgzc eag3qlw9"})
    games_description = soup.find_all("p",{"class":"description css-cput12 eag3qlw5"})
    prices = soup.find_all("div",{"class":"price-wrapper css-li4v8k eag3qlw4"})
    link_tags = soup.find_all("a", {"class": "card-header css-o171kl eag3qlw2"})
    games_links = [pre_url + link['href'] for link in link_tags if 'href' in link.attrs]
    
    for i in range(len(games_names)):
        game_name.append(games_names[i].text)
        game_description.append(games_description[i].text)
        genre.append(genres[i].text)
        price.append(prices[i].text)
        links.append(games_links[i].strip())
    page_num +=1
    print("page switched")
```

```python
extracted_price = [re.search(r"\d+,\d+", s).group() for s in price if re.search(r"\d+,\d+", s)]
```

```python
for link in links:
        web_result = requests.get(link)
        src = web_result.content
        soup = BeautifulSoup(src, "lxml")
        brand_info = soup.find("div", {"class": "brand-wrapper css-1f150rr e15c0rei0"})
        
        if brand_info:
            developers = brand_info.find("span", {"class": "brand developer"})
            if developers:
                developer.append(developers.text.replace("Developer:", "").strip())
            else:
                developer.append(None)
            types = brand_info.find_all("span")[-1] if brand_info.find_all("span") else None
            if types:
                type.append(types.text.replace("Type:", "").strip())
            else:
                type.append(None)
            meta_platform = soup.find('meta', property='og:platform')
            if meta_platform:
                platform.append(meta_platform['content'].strip("[]'\""))
            else:
                platform.append(None)
            availability = soup.find("p", class_="availability")
            if availability and "out of stock" in availability.text.strip().lower():
                stock_Status.append("Out of Stock")
            else:
                stock_Status.append("In Stock")
                
        else:
            developer.append(None)
            type.append(None)
```

```python
file_list = [game_name,game_description,genre,links,extracted_price,developer,type,platform,stock_Status]
exported = zip_longest(*file_list)
```

```python
with open("E:/CAT/webscarapping/task/WebScrapingResult.csv", "w", newline='', encoding='utf-8') as scrapResult: 
    wr = csv.writer(scrapResult)
    wr.writerow(["game name","game description","genre","link","price","developer","type","platform","stock status"])
    wr.writerows(exported)
```
